```python
teo_data = pd.read_csv("data_General.csv")
print(teo_data)
```

```python
LD_data = pd.read_csv("dataset_LD.csv")
print(LD_data)
```

```python
exp_data = pd.read_csv("data_LD_t.csv")
print(exp_data)
```

```python
print(exp_data[(exp_data["Z"]>8) & (exp_data["A"]>8)])
```

```python
merge_teo_exp = pd.merge(exp_data, teo_data, on=["Z","N"], how="inner")
print(merge_teo_exp)
```

```python
merge_ld_exp = pd.merge(exp_data, LD_data, on=["Z","N"], how="inner")
print(merge_ld_exp)
```

```python
merge_teo_exp["Bea_exp-Bea_pred"]=merge_teo_exp["BEA_x"]-merge_teo_exp["BEA_y"]
merge_teo_exp.plot.scatter(x="A_x",y="Bea_exp-Bea_pred", color="none", marker="o", edgecolor="black")
plt.show()
```

```python
merge_ld_exp["Bea_exp-Bea_pred"]=merge_ld_exp["BEA_x"]-merge_ld_exp["BEA_y"]
merge_ld_exp.plot.scatter(x="A_x",y="Bea_exp-Bea_pred", color="none",marker="^",edgecolor="blue")
plt.show()
```

```python
plt.scatter(x=merge_ld_exp["A_x"],y=merge_ld_exp["Bea_exp-Bea_pred"], color="none",marker="^",edgecolor="blue",s=5)
plt.scatter(x=merge_teo_exp["A_x"],y=merge_teo_exp["Bea_exp-Bea_pred"], color="none", marker="o", edgecolor="black",s=5)
plt.show()
```

```python
ws4_data = pd.read_csv("dataset_WS4.csv")
merge_ws4_exp = pd.merge(exp_data, ws4_data, on=["Z","N"], how="inner")
merge_ws4_exp["Bea_exp-Bea_pred"]=merge_ws4_exp["BEA_x"]-merge_ws4_exp["BEA_y"]
merge_ws4_exp.plot.scatter(x="A_x",y="Bea_exp-Bea_pred", color="none",marker="s",edgecolor="red",s=5)
plt.show()
```

```python
plt.scatter(x=merge_teo_exp["A_x"],y=merge_teo_exp["Bea_exp-Bea_pred"], color="none", marker="o", edgecolor="black",s=5)
plt.scatter(x=merge_ws4_exp["A_x"],y=merge_ws4_exp["Bea_exp-Bea_pred"], color="none",marker="s",edgecolor="red",s=5)
#plt.yscale("log")
plt.ylim()
plt.show()
```

```python
sly4_data = pd.read_csv("dataset_Sly4.csv")
merge_sly4_exp = pd.merge(exp_data, sly4_data, on=["Z","N"], how="inner")
merge_sly4_exp["Bea_exp-Bea_pred"]=merge_sly4_exp["BEA_x"]-merge_sly4_exp["BEA_y"]
plt.scatter(x=merge_teo_exp["A_x"],y=merge_teo_exp["Bea_exp-Bea_pred"], color="none", marker="o", edgecolor="black",s=5)
plt.scatter(x=merge_sly4_exp["A_x"],y=merge_sly4_exp["Bea_exp-Bea_pred"], color="none",marker="s",edgecolor="green",s=5)
plt.show()
```

```python
plt.scatter(x=merge_ld_exp["A_x"],y=merge_ld_exp["Bea_exp-Bea_pred"], color="none",marker="^",edgecolor="blue",s=5)
plt.scatter(x=merge_teo_exp["A_x"],y=merge_teo_exp["Bea_exp-Bea_pred"], color="none", marker="o", edgecolor="black",s=5)
plt.show()
```

```python
merge_sly4_exp["abs_bea_diff"]=np.abs(merge_sly4_exp["BEA_x"]-merge_sly4_exp["BEA_y"])
merge_ws4_exp["abs_bea_diff"]=np.abs(merge_ws4_exp["BEA_x"]-merge_ws4_exp["BEA_y"])
merge_ld_exp["abs_bea_diff"]=np.abs(merge_ld_exp["BEA_x"]-merge_ld_exp["BEA_y"])
merge_teo_exp["abs_bea_diff"]=np.abs(merge_teo_exp["BEA_x"]-merge_teo_exp["BEA_y"])
merge_sly4_exp["abs_bea_porc"]=np.abs(merge_sly4_exp["BEA_x"]-merge_sly4_exp["BEA_y"])/np.abs(merge_sly4_exp["BEA_x"])
merge_ws4_exp["abs_bea_porc"]=np.abs(merge_ws4_exp["BEA_x"]-merge_ws4_exp["BEA_y"])/np.abs(merge_ws4_exp["BEA_x"])
merge_ld_exp["abs_bea_porc"]=np.abs(merge_ld_exp["BEA_x"]-merge_ld_exp["BEA_y"])/np.abs(merge_ld_exp["BEA_x"])
merge_teo_exp["abs_bea_porc"]=np.abs(merge_teo_exp["BEA_x"]-merge_teo_exp["BEA_y"])/np.abs(merge_ld_exp["BEA_x"])
```

```python
plt.scatter(x=merge_teo_exp["A_x"],y=merge_teo_exp["abs_bea_porc"], color="none", marker="o", edgecolor="black",s=15,label="ANN",alpha=0.8)
plt.scatter(x=merge_ld_exp["A_x"],y=merge_ld_exp["abs_bea_porc"], color="none",marker="^",edgecolor="blue",s=15,label="LDM",alpha=0.35)
plt.yscale("log")
plt.ylim(10e-8,1)
plt.ylabel(r"$log(\frac{|BEA_{exp}-BEA_{pred}|}{BEA_{exp}}) \ {\rm (MeV)}$")
plt.xlabel("A")
plt.legend()
plt.tick_params(axis='x', which="both", direction='in')
plt.tick_params(axis='y', which="both", direction='in')
plt.grid(True)
plt.show()
```

```python
plt.scatter(x=merge_teo_exp["A_x"],y=merge_teo_exp["abs_bea_porc"], color="none", marker="o", edgecolor="black",s=5,label="ANN",alpha=0.8)
plt.scatter(x=merge_sly4_exp["A_x"],y=merge_sly4_exp["abs_bea_porc"], color="none",marker="s",edgecolor="green",s=5,label="Sly4",alpha=0.35)
plt.yscale("log")
plt.ylim(10e-8,1)
plt.ylabel(r"$log(\frac{|BEA_{exp}-BEA_{pred}|}{BEA_{exp}}) \ {\rm (MeV)}$")
plt.xlabel("A")
plt.legend()
plt.tick_params(axis='x', which="both", direction='in')
plt.tick_params(axis='y', which="both", direction='in')
plt.grid(True)
plt.show()
```

```python
plt.scatter(x=merge_teo_exp["A_x"],y=merge_teo_exp["abs_bea_porc"], color="none", marker="o", edgecolor="black",s=5,label="ANN",alpha=0.8)
plt.scatter(x=merge_ws4_exp["A_x"],y=merge_ws4_exp["abs_bea_porc"], color="none",marker="D",edgecolor="red",s=5,label="WS4",alpha=0.35)
plt.yscale("log")
plt.ylim(10e-8,1)
plt.ylabel(r"$log(\frac{|BEA_{exp}-BEA_{pred}|}{BEA_{exp}}) \ {\rm (MeV)}$")
plt.xlabel("A")
plt.legend()
plt.tick_params(axis='x', which="both", direction='in')
plt.tick_params(axis='y', which="both", direction='in')
plt.grid(True)
plt.show()
```

```python
ann_score = merge_teo_exp["abs_bea_diff"].mean()
ld_score = merge_ld_exp["abs_bea_diff"].mean()
ws4_score = merge_ws4_exp["abs_bea_diff"].mean()
sly4_score = merge_sly4_exp["abs_bea_diff"].mean()
print(" ANN   score  ", ann_score, "\n", "LD    score  ", ld_score, "\n", "WS4   score  ", ws4_score, "\n", "SLY4  score  ", sly4_score, "\n")
```
