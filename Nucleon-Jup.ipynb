```python
#desviacion absoluta y procentual de la energia de enlace por nucleon
merge_sly4_exp["abs_bea_diff"]=np.abs(merge_sly4_exp["BEA_x"]-merge_sly4_exp["BEA_y"])
merge_ws4_exp["abs_bea_diff"]=np.abs(merge_ws4_exp["BEA_x"]-merge_ws4_exp["BEA_y"])
merge_ld_exp["abs_bea_diff"]=np.abs(merge_ld_exp["BEA_x"]-merge_ld_exp["BEA_y"])
merge_teo_exp["abs_bea_diff"]=np.abs(merge_teo_exp["BEA_x"]-merge_teo_exp["BEA_y"])
merge_sly4_exp["abs_bea_porc"]=np.abs(merge_sly4_exp["BEA_x"]-merge_sly4_exp["BEA_y"])/np.abs(merge_sly4_exp["BEA_x"])
merge_ws4_exp["abs_bea_porc"]=np.abs(merge_ws4_exp["BEA_x"]-merge_ws4_exp["BEA_y"])/np.abs(merge_ws4_exp["BEA_x"])
merge_ld_exp["abs_bea_porc"]=np.abs(merge_ld_exp["BEA_x"]-merge_ld_exp["BEA_y"])/np.abs(merge_ld_exp["BEA_x"])
merge_teo_exp["abs_bea_porc"]=np.abs(merge_teo_exp["BEA_x"]-merge_teo_exp["BEA_y"])/np.abs(merge_ld_exp["BEA_x"])

#desviacion rms en energia por nucleon
rsmd_sly4_exp = 0
for i in merge_sly4_exp["abs_bea_diff"]: rsmd_sly4_exp = i*i
rsmd_sly4_exp = np.sqrt(rsmd_sly4_exp/merge_sly4_exp["abs_bea_diff"].size)

rsmd_ws4_exp = 0
for i in merge_ws4_exp["abs_bea_diff"]: rsmd_ws4_exp = i*i
rsmd_ws4_exp = np.sqrt(rsmd_ws4_exp/merge_ws4_exp["abs_bea_diff"].size)

rsmd_ld_exp = 0
for i in merge_ld_exp["abs_bea_diff"]: rsmd_ld_exp = i*i
rsmd_ld_exp = np.sqrt(rsmd_ld_exp/merge_ld_exp["abs_bea_diff"].size)

rsmd_teo_exp = 0
for i in merge_teo_exp["abs_bea_diff"]: rsmd_teo_exp = i*i
rsmd_teo_exp = np.sqrt(rsmd_teo_exp/merge_teo_exp["abs_bea_diff"].size)

#desviacion rms entre masas nucleares
merge_sly4_exp["abs_bea_massdiff"]=np.abs(merge_sly4_exp["BEA_x"]*merge_sly4_exp["A_x"] - merge_sly4_exp["BEA_y"]*merge_sly4_exp["A_x"])
mass_rsmd_sly4_exp = 0
for i in merge_sly4_exp["abs_bea_massdiff"]: mass_rsmd_sly4_exp = i*i
mass_rsmd_sly4_exp = np.sqrt(mass_rsmd_sly4_exp/merge_sly4_exp["abs_bea_massdiff"].size)

merge_ws4_exp["abs_bea_massdiff"]=np.abs(merge_ws4_exp["BEA_x"]*merge_ws4_exp["A_x"] - merge_ws4_exp["BEA_y"]*merge_ws4_exp["A_x"])
mass_rsmd_ws4_exp = 0
for i in merge_ws4_exp["abs_bea_massdiff"]: mass_rsmd_ws4_exp = i*i
mass_rsmd_ws4_exp = np.sqrt(mass_rsmd_ws4_exp/merge_ws4_exp["abs_bea_massdiff"].size)

merge_teo_exp["abs_bea_massdiff"]=np.abs(merge_teo_exp["BEA_x"]*merge_teo_exp["A_x"] - merge_teo_exp["BEA_y"]*merge_teo_exp["A_x"])
mass_rsmd_teo_exp = 0
for i in merge_teo_exp["abs_bea_massdiff"]: mass_rsmd_teo_exp = i*i
mass_rsmd_teo_exp = np.sqrt(mass_rsmd_teo_exp/merge_teo_exp["abs_bea_massdiff"].size)

merge_ld_exp["abs_bea_massdiff"]=np.abs(merge_ld_exp["BEA_x"]*merge_ld_exp["A_x"] - merge_ld_exp["BEA_y"]*merge_ld_exp["A_x"])
mass_rsmd_ld_exp = 0
for i in merge_ld_exp["abs_bea_massdiff"]: mass_rsmd_ld_exp = i*i
mass_rsmd_ld_exp = np.sqrt(mass_rsmd_ld_exp/merge_ld_exp["abs_bea_massdiff"].size)


```
